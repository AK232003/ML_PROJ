```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, roc_auc_score
```

```python
train_df = pd.read_csv("./Preprocessed.csv")
test_df = pd.read_csv("./Test_Preprocessed.csv")
```

```python
train_df.head()
```

```python
test_df.drop('Unnamed: 0', inplace=True, axis =1)
```

```python
train_df.drop('Unnamed: 0', inplace=True, axis =1)
```

```python
Ytrain = train_df['isFraud']
```

```python
train_df.drop('isFraud', inplace=True, axis =1)
```

```python
param_grid = {
    'var_smoothing': np.logspace(0,-9, num=100)
}
```

var_smoothingfloat, default=1e-9

    Portion of the largest variance of all features that is added to variances for calculation stability.

the aim of smoothing is to give a general idea of relatively slow changes of value with little attention paid to the close matching of data values, while curve fitting concentrates on achieving as close a match as possible.

```python
gnbModel = GridSearchCV(estimator=GaussianNB(), param_grid=param_grid, verbose=1, cv=10, n_jobs=-1)
gnbModel.fit(train_df, Ytrain)
print(gnbModel.best_params_)
```

```python
ypredtrain = gnbModel.predict(train_df)
```

```python
print(f1_score(Ytrain, ypredtrain))
```

```python
print(roc_auc_score(Ytrain, ypredtrain))
```

```python
ypred = gnbModel.predict(test_df)
```

```python
for i in ypred:
    print(i)
```

```python
CSV4 = pd.DataFrame(ypred)
file = CSV4.to_csv("PredictionsGNB.csv")
```
