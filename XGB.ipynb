```python
import numpy as np
import pandas as pd
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
```

```python
train_df = pd.read_csv("./PreprocessedTrain.csv")
test_df = pd.read_csv("./PreprocessedTest.csv")
```

```python
train_df.head()
```

```python
test_df.drop('Unnamed: 0', inplace=True, axis =1)
```

```python
train_df.drop('Unnamed: 0', inplace=True, axis =1)
```

```python
Ytrain = train_df['isFraud']
```

```python
train_df.drop('isFraud', inplace=True, axis =1)
```

```python
train_df.head()
```

### Parameters

Mainly these parameters are to avoid overfitting

n_estimators — The number of runs XGBoost will try to learn


learning_rate — learning speed

max_depth - Max Depth of the tree.

colsample_bytree - defines what percentage of features ( columns ) will be used for building each tree.

subsample - Fraction of the dataset that will be used for training

gpu_hist employs gradient based boosting so it will be good to employ it. (I have good gpu XD)

```python
param_grid = {
    'n_estimators':[2000], 
    'max_depth':[12], 
    'learning_rate':[0.2], 
    'subsample':[0.8],
    'colsample_bytree':[0.4],  
    'eval_metric':['auc'],
    'tree_method':['gpu_hist']
    }
```

```python
xgbModel = GridSearchCV(estimator=XGBClassifier(), param_grid=param_grid, verbose=1, cv=5, n_jobs=-1)
xgbModel.fit(train_df, Ytrain)
```

```python
print(xgbModel.best_params_)
```

```python
print(xgbModel.best_score_)
```

```python
ypred = xgbModel.predict(test_df)
```

```python
for i in ypred:
    print(i)
```

```python
CSV4 = pd.DataFrame(ypred)
file = CSV4.to_csv("PredictionsXGB.csv")
```
